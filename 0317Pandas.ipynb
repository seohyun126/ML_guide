## 결손 데이터 처리하기

```python
titanic_df.isna().sum()
```

```python
titanic_df['Cabin']=titanic_df['Cabin'].fillna('C000')
#titanic_df['Cabin'].fillna('C000',inplace=True)<- 이렇게 써도 됨
titanic_df.head(5)
#결손 데이터 대체하기
```

## 데이터 가공: apply lambda

```python
lambda_thirth=lambda x: x**3
print(lambda_thirth(5))
```

```python
a=[1,3,5]
squares=map(lambda x:x**3,a)
list(squares)
```

```python
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x:len(x))
titanic_df[['Name','Name_len']].head(5)
```

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(5)
#이때 if 식보다 if 가 true일 때 반환값 먼저 쓰기
```

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df[['Age','Age_cat']].head(5)
```
